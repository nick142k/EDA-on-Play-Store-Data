## 3. ***Data Wrangling***

```python
data.head()
```

```python
# In this  cell I will define a funtion called remove_emojis, for my first column which is App using regular expression.

import re

# Define the regular expression pattern and using Regular expression unicode's for filter App column.

pattern = r"[" \
          u"\U0001F600-\U0001F64F" \
          u"\U0001F300-\U0001F5FF" \
          u"\U0001F680-\U0001F6FF" \
          u"\U0001F1E0-\U0001F1FF" \
          "]"

# Function to remove emojis and flags
def remove_emojis(text):
    return re.sub(pattern, '', text)

# Apply the function to the 'App' column
data['App'] = data['App'].apply(lambda x: remove_emojis(x))
```

```python
# In this cell  I have cleaned three columns which has inconsitent data present in the data, again using regular expression. 
# But before that let's first remvove  the 'Unnamed: 0' column which has no use for in the dataframe using Python drop inbuilt funtion.

data.drop(columns= 'Unnamed: 0',inplace = True) 



data['Price'] = data['Price'].replace(r'[$]','',regex=True).astype(float)
data['Installs']=data['Installs'].replace(r'[+,]','',regex=True)
data['Installs']=data['Installs'].astype(int)
```

### From now I am going to handle null values available in the columns


### Let's first explore the rating columns which have total 1463 rows with NaN values. 

```python
#Exploring the Rating column to analyze the ouliers
sns.boxplot(data = data['Rating'])
plt.title('Box plot of rating column')
```

```python
# Exploring the Rating column for itentify the column follows the Normal distribution or not 

sns.kdeplot(data=data[['Rating']])
plt.title('Kernal density plot for analysing  Normal Distribution')
```

```python
# From my above analysis, I have seen that the Rating column is has very much outliers and it was also not following normal distribution so in the below code i have imputer the null values with median of the column.

replaceing_nan_with_mean_in_rating = data.groupby('Category')['Rating'].transform('median')
data['Rating'].fillna(replaceing_nan_with_mean_in_rating,inplace=True)
data['Rating']
```

### Manipulating the 'Size' column has also has total  1227 NuN value.

```python
# Exploring the size column for itentify the outliers

sns.boxplot(data=data['Size'])
plt.title('Box plot of size column')
```

```python
# Exploring the size column for itentify the column follows the Normal distribution or not 

sns.kdeplot(data=data['Size'])
plt.title('Kernal density plot for checking Normal distibution')
```

```python
replacing_mean_with_median_in_size=data.groupby('Category')['Size'].transform('median')# From my above analysis, I have seen that the size column is has very much outliers and it was also not following normal distribution so in the below code i have imputer the null values with median of the column.

data['Size'].fillna(replacing_mean_with_median_in_size,inplace=True)
```
